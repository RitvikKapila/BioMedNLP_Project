```python
import os

folder_name = 'Patients_Datewise_Texts/'
os.makedirs(folder_name, exist_ok=True)

directory_path = 'selected_patients/'
csv_files = glob.glob(os.path.join(directory_path, '*.csv'))


for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    df = df.sort_values(['CHARTDATE','CHARTTIME'])
    texts = df[['CHARTDATE', 'TEXT']]
    result_df = texts.groupby('CHARTDATE')['TEXT'].agg(lambda x: ' '.join(x)).reset_index()
    result_df.to_csv(folder_name + 'text_' + str(csv_file.split('/')[-1]), index=False)
```

```python
def compare_dates(row, given_date):
    return 1 if row['CHARTDATE'] >= given_date else 0
```

```python
folder_name = 'Patients_Datewise_Texts_With_First_Occurance/'
os.makedirs(folder_name, exist_ok=True)

directory_path = 'Patients_Datewise_Texts/'
csv_files = glob.glob(os.path.join(directory_path, '*.csv'))

ctr = 0

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    try:
        date = df[df['TEXT'].str.contains('pulmonary embolism', case=False)]['CHARTDATE'].iloc[0]
        df['FIRST_OCCURANCE'] = df.apply(compare_dates, args=(date, ), axis=1)
        df.to_csv(folder_name + 'fo_' + str(csv_file.split('/')[-1]), index=False)
    except:
        ctr += 1

print(ctr)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("togethercomputer/LLaMA-2-7B-32K")
model = AutoModelForCausalLM.from_pretrained("togethercomputer/LLaMA-2-7B-32K", trust_remote_code=True, torch_dtype=torch.float16)
```

```python
device = "cuda:0"
model.to(device)
```

```python
# input_context = "For this clinical notes: " + df["TEXT"].iloc[0] + ". Predict if this patient has pulmonary embolism based on the given notes above. Give an yes/no answer" 
input_context = "Does the sun rise in the east or west? Think like a normal person"
input_ids = tokenizer.encode(input_context, return_tensors="pt")
input_ids = input_ids.to(device)
output = model.generate(input_ids, max_length=128, temperature=0.7)
output_text = tokenizer.decode(output[0], skip_special_tokens=True)
```
